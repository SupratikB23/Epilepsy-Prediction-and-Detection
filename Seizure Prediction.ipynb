```python
import os
import numpy as np
import mne
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import Input, Dense, Dropout, MultiHeadAttention, Add, LayerNormalization, Flatten, Reshape
from sklearn.utils import class_weight
from sklearn.preprocessing import StandardScaler
from scipy.stats import entropy
from google.colab import drive
import tensorflow as tf

drive.mount('/content/drive')

SAMPLING_RATE = 512
DOWNSAMPLED_RATE = 128
WINDOW_SIZE_SEC = 5
WINDOW_SIZE = WINDOW_SIZE_SEC * DOWNSAMPLED_RATE
CHANNELS = 34

dataset_path = '/content/drive/My Drive/Colab Notebooks/IEEE EMBS Internship/EEG Dataset'

# Define Seizure Log
def time_to_seconds(t_str):
    h, m, s = map(int, t_str.split(':'))
    return h * 3600 + m * 60 + s

seizure_log = {
    "PN00-1.edf": ["19:39:33", ["19:58:36"]],
    "PN00-2.edf": ["02:18:17", ["02:38:37"]],
    "PN00-3.edf": ["18:15:44", ["18:28:29"]],
    "PN00-4.edf": ["20:51:43", ["21:08:29"]],
}

manual_seizure_times = {
    fname: [time_to_seconds(ot) - time_to_seconds(start)]
    for fname, (start, ots) in seizure_log.items()
    for ot in ots
}

# EDF Processing & Feature
def extract_features(segment):
    features = []
    for ch_data in segment:
        mean = np.mean(ch_data)
        std = np.std(ch_data)
        var = np.var(ch_data)
        max_val = np.max(ch_data)
        min_val = np.min(ch_data)
        ptp_val = np.ptp(ch_data)
        ent = entropy(np.abs(np.fft.rfft(ch_data)) + 1e-8)
        features.extend([mean, std, var, max_val, min_val, ptp_val, ent])
    return np.array(features)

def preprocess_and_segment(file_path):
    file_name = os.path.basename(file_path)
    raw = mne.io.read_raw_edf(file_path, preload=True, verbose=False)
    raw.pick_channels(raw.ch_names[:CHANNELS])
    raw.resample(DOWNSAMPLED_RATE)
    data = raw.get_data()
    seizure_onsets = manual_seizure_times.get(file_name, [])

    X, y = [], []
    total_samples = data.shape[1]
    n_windows = total_samples // WINDOW_SIZE

    for i in range(n_windows):
        start = i * WINDOW_SIZE
        end = start + WINDOW_SIZE
        segment = data[:, start:end]
        if segment.shape[1] < WINDOW_SIZE:
            continue

        time_sec = start / DOWNSAMPLED_RATE
        label = int(any((onset - 300) <= time_sec <= onset for onset in seizure_onsets))
        features = extract_features(segment)
        X.append(features)
        y.append(label)

    return np.array(X), np.array(y)

# Load Data
edf_files = sorted([f for f in os.listdir(dataset_path) if f.endswith('.edf')])
train_files = edf_files[:4]
test_file = edf_files[4]

X_raw, y_raw = [], []
for f in train_files:
    path = os.path.join(dataset_path, f)
    x, y = preprocess_and_segment(path)
    X_raw.extend(x)
    y_raw.extend(y)

X_raw = np.array(X_raw)
y_raw = np.array(y_raw)

X_train, X_val, y_train, y_val = train_test_split(
    X_raw, y_raw, test_size=0.1, random_state=42, stratify=y_raw
)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)

X_train = X_train[..., np.newaxis]
X_val = X_val[..., np.newaxis]

print(f"‚úÖ Training samples: {len(X_train)} (preictal: {sum(y_train)})")
print(f"‚úÖ Validation samples: {len(X_val)} (preictal: {sum(y_val)})")

# Model (Attention-Augmented MLP)
def attention_block(inputs, num_heads=4, key_dim=8):
    attention = MultiHeadAttention(num_heads=num_heads, key_dim=key_dim)(inputs, inputs)
    attention = Dropout(0.2)(attention)
    attention = Add()([inputs, attention])
    attention = LayerNormalization()(attention)
    return attention

input_layer = Input(shape=(X_train.shape[1], 1))
x = Flatten()(input_layer)
x = Dense(128, activation='relu')(x)
x = Dropout(0.3)(x)
x = Dense(64, activation='relu')(x)
x = Dropout(0.2)(x)
x = Reshape((1, -1))(x)
x = attention_block(x)
x = Flatten()(x)
x = Dense(64, activation='relu')(x)
output = Dense(1, activation='sigmoid')(x)

model = Model(inputs=input_layer, outputs=output)
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

weights = class_weight.compute_class_weight(class_weight='balanced', classes=np.unique(y_train), y=y_train)
class_weights = {0: weights[0], 1: weights[1]}

history = model.fit(
    X_train, y_train,
    epochs=30,
    batch_size=16,
    validation_data=(X_val, y_val),
    class_weight=class_weights
)

val_acc = history.history['val_accuracy'][-1]
print(f"\n‚úÖ Final Validation Accuracy: {val_acc:.4f}")

model.save('/content/drive/My Drive/Colab Notebooks/IEEE EMBS Internship/seizure_predictor_attention.h5')

# Accuracy Graph
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.title('Training vs Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()

# Predict on Final Test File
test_path = os.path.join(dataset_path, test_file)
X_test, y_test = preprocess_and_segment(test_path)
X_test = scaler.transform(X_test)
X_test = X_test[..., np.newaxis]
model = load_model('/content/drive/My Drive/Colab Notebooks/IEEE EMBS Internship/seizure_predictor_attention.h5')

print(f"\nüß™ Test file: {test_file}")
print(f"Total windows: {len(X_test)}, Preictal windows (True): {sum(y_test)}")

probs = model.predict(X_test)
preds = (probs >= 0.4).astype(int).flatten()
print(f"Predicted Preictal: {sum(preds)}")

acc = np.mean(preds == y_test)
print(f"‚úÖ Accuracy: {acc:.4f}")

for i, (p, t) in enumerate(zip(preds, y_test)):
    if p == 1:
        print(f"‚ö†Ô∏è Preictal Detected at {i * 5 // 60} min (Prob: {probs[i][0]:.3f}) | True: {t}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_detection_timings_lines(auto_probs, auto_preds, mlp_probs, mlp_preds, att_probs, att_preds):
    window_size_min = 5 / 60

    def get_minute_prob_pairs(preds, probs):
        minutes = []
        probabilities = []
        for i, (p, prob) in enumerate(zip(preds, probs)):
            time_min = i * window_size_min
            if 15 <= time_min <= 19 and p == 1:
                minutes.append(time_min)
                probabilities.append(np.clip(prob, 0, 1))
        return minutes, probabilities

    auto_mins, auto_vals = get_minute_prob_pairs(auto_preds, auto_probs)
    mlp_mins, mlp_vals = get_minute_prob_pairs(mlp_preds, mlp_probs)
    att_mins, att_vals = get_minute_prob_pairs(att_preds, att_probs)

    plt.figure(figsize=(10, 6))
    plt.plot(auto_mins, auto_vals, label='Autoencoder', marker='o', color='red')
    plt.plot(mlp_mins, mlp_vals, label='MLP (CNN-BiLSTM)', marker='s', color='blue')
    plt.plot(att_mins, att_vals, label='Attention Model', marker='^', color='green')

    plt.title("Preictal Detection Probabilities")
    plt.xlabel("Time (minutes)")
    plt.ylabel("Anomaly Detected MSE")
    plt.ylim(0, 1)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

plot_detection_timings_lines(mse, autoencoder_preds, mlp_probs.flatten(), mlp_preds, att_probs.flatten(), att_preds)
```
